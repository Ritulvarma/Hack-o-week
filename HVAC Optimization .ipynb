```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
```

```python
# Sample dataset
data = pd.DataFrame({
    'temperature': [22, 25, 28, 30, 35, 24, 27, 32, 29, 26],
    'occupancy': [5, 10, 15, 20, 25, 8, 12, 18, 14, 9],
    'cooling_needed': [0, 0, 1, 1, 1, 0, 1, 1, 1, 0]
})

data
```

```python
X = data[['temperature', 'occupancy']]
y = data['cooling_needed']

model = DecisionTreeClassifier()
model.fit(X, y)

print("Model Trained Successfully!")
```

```python
temp = int(input("Enter Temperature: "))
occupancy = int(input("Enter Occupancy: "))

prediction = model.predict([[temp, occupancy]])

if prediction[0] == 1:
    print("Cooling Required ❄️")
else:
    print("Cooling Not Required 🌿")
```

```python
# Create dummy zone data (5x5 lab zones)
zone_data = np.random.randint(0, 2, (5, 5))

plt.figure()
sns.heatmap(zone_data, annot=True)
plt.title("Zone-wise Cooling Heatmap")
plt.show()
```

```python
# Scatter plot
plt.figure()
sns.scatterplot(x='temperature', y='occupancy', hue='cooling_needed', data=data)
plt.title("Cooling Decision Based on Temperature & Occupancy")
plt.show()
```
