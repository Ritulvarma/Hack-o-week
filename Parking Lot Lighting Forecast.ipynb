```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)

data = pd.DataFrame({
    'vehicles': np.random.randint(0, 100, 50),
    'light_usage': np.random.randint(20, 200, 50)
})

data.head()
```

```python
X = data[['vehicles']]
y = data['light_usage']

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
```

```python
vehicle_input = 60

predicted_usage = model.predict(poly.transform([[vehicle_input]]))
print("Predicted Light Usage:", predicted_usage[0])
```

```python
plt.figure()
plt.bar(data['vehicles'], data['light_usage'])
plt.xlabel("Vehicle Count")
plt.ylabel("Light Usage")
plt.title("Parking Lot Lighting Usage")
plt.show()
```

```python
# Define threshold (example)
threshold = data['light_usage'].mean() + 2 * data['light_usage'].std()

anomalies = data[data['light_usage'] > threshold]

print("Anomalies Detected:")
print(anomalies)
```

```python
plt.figure()

plt.bar(data['vehicles'], data['light_usage'], label='Normal')

plt.scatter(anomalies['vehicles'], anomalies['light_usage'], 
            color='red', label='Anomaly')

plt.legend()
plt.title("Anomaly Detection in Lighting Usage")
plt.show()
```
